```python
import cv2
import pandas as pd
import os

# Load the CSV file containing video names and labels
csv_path = "newshoplifting.csv"
df = pd.read_csv(csv_path)

# Filter videos labeled as "shoplifting" and "normal"
shoplifting_videos = df[df['label'] == 1]['videoname'].tolist()
normal_videos = df[df['label'] == 0]['videoname'].tolist()

def preprocess_videos(video_list, output_folder):
    os.makedirs(output_folder, exist_ok=True)
    for video_name in video_list:
        video_path = f"dataset/{video_name}.mp4"
        output_path = f"{output_folder}/{video_name}/"
        os.makedirs(output_path, exist_ok=True)
        # Use OpenCV to read the video
        cap = cv2.VideoCapture(video_path)
        frame_count = 0
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            # Process the frame (e.g., resize, convert to grayscale)
            processed_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            # Save the processed frame
            cv2.imwrite(f"{output_path}/frame_{frame_count}.jpg", processed_frame)
            frame_count += 1
        cap.release()


# Preprocess shoplifting and normal videos
output_folder_shoplifting = "outputshoplifting"
output_folder_normal = "outputnormal"
preprocess_videos(shoplifting_videos, output_folder_shoplifting)
preprocess_videos(normal_videos, output_folder_normal)
```
